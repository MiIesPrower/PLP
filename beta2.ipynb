```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters for the Beta distributions
alpha1, beta1 = 0.25, 4.0
alpha2, beta2 = 4.0, 0.25

# Number of samples
n_samples = 1000

# Generate samples from each Beta distribution
samples1 = beta.rvs(alpha1, beta1, size=n_samples//2)
samples2 = beta.rvs(alpha2, beta2, size=n_samples//2)

# Combine the samples
samples = np.concatenate([samples1, samples2])

# Plot the distribution
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')

# Plot the theoretical density functions
x = np.linspace(0, 1, 1000)
plt.plot(x, 0.5 * beta.pdf(x, alpha1, beta1), 'r-', lw=2, label='Beta(0.25, 4.0)')
plt.plot(x, 0.5 * beta.pdf(x, alpha2, beta2), 'b-', lw=2, label='Beta(4.0, 0.25)')
plt.legend(loc='best')
plt.title('Mixture of Two Beta Distributions')
plt.show()

```
